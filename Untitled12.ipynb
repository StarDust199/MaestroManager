```python
import os
from PIL import Image

def generate_yolo_labels(image_folder, label_folder, class_id):
    # Upewnij się, że folder na etykiety istnieje
    os.makedirs(label_folder, exist_ok=True)

    # Przejdź przez wszystkie obrazy w folderze
    for image_name in os.listdir(image_folder):
        if image_name.endswith('.jpg') or image_name.endswith('.png'):
            image_path = os.path.join(image_folder, image_name)
            label_path = os.path.join(label_folder, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))

            # Otwórz obraz, aby uzyskać jego wymiary
            with Image.open(image_path) as img:
                img_width, img_height = img.size

            # Współrzędne bounding boxa (cały obraz)
            x_center = 0.5  # Środek obrazu w osi X
            y_center = 0.5  # Środek obrazu w osi Y
            width = 1.0     # Szerokość bounding boxa (cały obraz)
            height = 1.0    # Wysokość bounding boxa (cały obraz)

            # Zapisz etykietę w formacie YOLO
            with open(label_path, 'w') as f:
                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

# Przykład użycia
generate_yolo_labels(
    image_folder='/content/drive/MyDrive/RoadsignsDB/Test/Images',
    label_folder='/content/drive/MyDrive/RoadsignsDB/Test/Labels',
    class_id=0  # ID klasy dla znaku drogowego (dostosuj do swoich danych)
)
```
